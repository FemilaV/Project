```python
re.findall('\d{2} [a-zA-Z]{3} \d{2}',df.disaster[0])
```

```python
# extracting from_date from the disaster column
from_date = []
for i in df.disaster:
    a = re.findall('\d{2} [a-zA-Z]{3} \d{2}',i)
    from_date.append(a[0])
```

```python
from_date
```

```python
df1 = pd.DataFrame(from_date)
df1
```

```python
df1.columns = ['from_date']
```

```python
re.findall('[a-zA-Z]{3} \d{2}, \d{4}',df.date[0])
```

```python
# extracting only the to-date from the date column without time
to_DATE = []
for i in df.date:
    a = re.findall('[a-zA-Z]{3} \d{2}, \d{4}',i)
    to_DATE.append(a)
```

```python
to_DATE
```

```python
df2 = pd.DataFrame(to_DATE)
df2
```

```python
df2.columns=['to_date']
```

```python
df2
```

```python
# concatenating the from-date,to-date and original dataset
data = pd.concat([df1,df2,df],axis=1)
```

```python
data
```

```python
# dropping unwanted columns
df4 = data.drop(['date'],axis=1,inplace=True)
```

```python
data
```

```python
# removing the dates from the disaster name
data["disaster"]= data["disaster"].str.split("from" , n = 1, expand = True)
```

```python
data
```

```python
data["disaster"]= data["disaster"].str.split("at", n = 1, expand = True)
```

```python
# converting dates to datetime format
data['from_date'] = pd.to_datetime(data['from_date'])
data['to_date'] = pd.to_datetime(data['to_date'])
```
